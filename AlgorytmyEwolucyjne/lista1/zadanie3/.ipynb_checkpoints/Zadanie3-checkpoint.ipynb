```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def initial_probability_vector(d):
    return np.ones(d) * 0.5

def binary_random(p):
    if np.random.random() < p:
        return 1
    return 0

def random_individual(p, d):
    x = np.zeros(d)
    for i in range(d):
        x[i] = binary_random(p[i])
    return x

def random_population(p, d, n):
    P = np.zeros((n, d))
    for i in range(n):
        P[i] = random_individual(p, d)
    return P

def population_evaluation(P, F, n):
    res = np.zeros(n)
    for i in range(n):
        res[i] = F(P[i])
    return res

def pbil( F, n, d, theta1, theta2, theta3 ):
    VP = initial_probability_vector(d)
    POP = random_population(VP, d, n)
    
    for i in range(1000):
        E = population_evaluation(POP, F, n)
        bi = E.argmax() #best individual
        VP = VP * (1.0 - theta1) + POP[bi] * theta1
        
        print(i)
        if i % 10 == 0:
            print("best individual:", bi, ", E[bi]:", E[bi])
            print("prob vector:{}", VP)
        
        for k in range(d):
            if np.random.random() < theta2:
                VP[k] = VP[k] * (1.0 - theta3) + binary_random(0.5) * theta3
        POP = random_population(VP, d, n)
        
    return VP
```

```python
classificationrulesfile = open("ClassificationRules.txt")
imagerawfile = open("ImageRawReduced.txt")
imageexpfile = open("ImageExpertReduced.txt")

rozmiarobraz = 9350
IEXP = np.empty((rozmiarobraz))
IRAW = np.empty((3, rozmiarobraz))
CLAS = np.empty((266, rozmiarobraz))

l1 = imageexpfile.readline()
IEXP = np.array(list(map(float, l1.split())))

print(IEXP.shape)
print(IEXP)

linenum = 0
for line in imagerawfile:
    IRAW[linenum, :] = np.array(list(map(float, line.split())))
    linenum += 1

linenum = 0
for line in classificationrulesfile:
    CLAS[linenum, :] = np.array(list(map(float, line.split())))
    linenum += 1
```

```python
def evaluate(V):
    RES = np.empty((rozmiarobraz))
    T = CLAS * V.reshape(266, 1) #wtedy wyzerowane zostaja rzedy ktorych nie bierzemy
    T2 = T.astype(np.int64)
    
    #print("V:", V, " nonzero:", np.count_nonzero(V))
    #print("T:", T, "shape:", T.shape)
    #print("T2:", T2, "shape:", T2.shape)
    
    for i in range(rozmiarobraz):
        v, c = np.unique(T2[:,i], return_counts=True)
        mostfrequentval = 1
        mostfrequentcount = 0
        for j in range(v.shape[0]):
            if c[j] > mostfrequentcount and v[j] != 0:
                mostfrequentcount = c[j]
                mostfrequentval = v[j]
        RES[i] = mostfrequentval
    
    #print("RES", RES)
    return np.count_nonzero(RES == IEXP.astype(np.int64))
```

```python
vres = pbil(evaluate, 20, 266, 0.1, 0.01, 0.1)
```

```python
print(vres)
```

```python
RANDOMP = random_population(vres, 266, 100)
PEVAL = population_evaluation(RANDOMP, evaluate, 100)
bestidx = PEVAL.argmax()
bestindividual = RANDOMP[bestidx]
print( PEVAL[bestidx] )
print(bestindividual)
print(IRAW)
```

```python
def calculatecolor(V):
    RES = np.empty((rozmiarobraz))
    T = CLAS * V.reshape(266, 1) #wtedy wyzerowane zostaja rzedy ktorych nie bierzemy
    T2 = T.astype(np.int64)
    
    for i in range(rozmiarobraz):
        v, c = np.unique(T2[:,i], return_counts=True)
        mostfrequentval = 1
        mostfrequentcount = 0
        for j in range(v.shape[0]):
            if c[j] > mostfrequentcount and v[j] != 0:
                mostfrequentcount = c[j]
                mostfrequentval = v[j]
        RES[i] = mostfrequentval
        
    return RES

bestcolors = calculatecolor(bestindividual)
print(calculatecolor(bestindividual))
```
